```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import scipy.stats as ss
import math
```

```python
wage2 = pd.read_stata('WAGE2.dta')
wage2 = wage2.dropna(subset=['wage'])
wage2.drop(wage2[wage2['wage'] < 1].index, inplace=True)
wage2.head()
```

```python
wage2['lwage'] = np.log(wage2['wage'])
educ = wage2['educ'].tolist()
exper = wage2['exper'].tolist()
educExper = []

for i in range(0,len(educ)):
    educExper.append(educ[i] * exper[i])
    
wage2['educExper'] = educExper
wage2
```

```python
x = wage2[['educ', 'exper', 'educExper']]
y = wage2[['lwage']]
x = sm.add_constant(x)
reg = sm.OLS(y,x).fit()
print(reg.summary())
```

```python
exper10 = wage2['exper']-10
exper10 = exper10.tolist()
edex10 = []

for i in range(0,len(exper10)):
    edex10.append(educ[i]*exper10[i])
    
wage2['edex10'] = edex10
```

```python
x = wage2[['educ', 'exper', 'edex10']]
y = wage2[['lwage']]
x = sm.add_constant(x)
reg = sm.OLS(y,x).fit()
print(reg.summary())
```

```python
gpa2 = pd.read_stata('GPA2_SMALL.dta')
gpa2.dropna(subset=['hsize'], inplace=True)
gpa2
```

```python
gpa2['hsizesq'] = gpa2['hsize'] ** 2
```

```python
x = gpa2[['hsize', 'hsizesq']]
y = gpa2[['sat']]
x = sm.add_constant(x)
reg = sm.OLS(y,x).fit()
print(reg.summary())
```

```python
gpa2['lsat'] = np.log(gpa2['sat'])
```

```python
x = gpa2[['hsize', 'hsizesq']]
y = gpa2[['lsat']]
x = sm.add_constant(x)
reg = sm.OLS(y,x).fit()
print(reg.summary())
```
